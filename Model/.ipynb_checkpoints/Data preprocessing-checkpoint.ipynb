# Merging of two seperate wine data csv files

```python
import pandas as pd 
import csv
```

```python
data_red =  pd.read_csv('winequality-red.csv', sep=';')
data_white =  pd.read_csv('winequality-white.csv', sep=';')
```

```python
data_red["wine type"]='red'
data_white["wine type"]='white'
```

```python
data_set=data_red.append(data_white)
```

```python
data_set.columns
```

```python
data_set=data_set.reindex(columns=['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol', 'wine type', 'quality'])
data_set
```

```python
data_set.to_csv('data_set.csv',index=False)
```
