# Merging of two seperate wine data csv files

```python
import pandas as pd 
import csv
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
data_red =  pd.read_csv('winequality-red.csv', sep=';')
data_white =  pd.read_csv('winequality-white.csv', sep=';')
```

```python
data_red["wine type"] = 1
data_white["wine type"]= 0
```

```python
data_set=data_red.append(data_white)
```

```python
data_set.columns
```

```python
data_set=data_set.reindex(columns=['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol', 'wine type', 'quality'])
data_set
```

```python
data_set.to_csv('data_set.csv',index=False)
```

```python
data_set.describe()
```

```python
sns.boxplot(data_set['residual sugar'])
```

```python
z = np.abs(stats.zscore(data_set))
# print(z)
data_set = data_set[(z < 3).all(axis=1)]
data_set.shape
```

```python
plt.subplots(figsize=(15, 10))
sns.heatmap(data_set.corr(), annot = True, cmap = 'coolwarm')
```

```python
data_set['quality'].value_counts()
```

```python
# Define features X
X = np.asarray(data_set.iloc[:,:-1]) 
# Define target y
y = np.asarray(data_set['quality'])
```

```python
from sklearn import preprocessing
X = preprocessing.StandardScaler().fit(X).transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=0)
print ('Train set:', X_train.shape, y_train.shape)
print ('Test set:', X_test.shape, y_test.shape)
```
